```python
def sum_distance(from_num, to_num):  
    # Проверяем, не больше ли первое число второго  
    if from_num > to_num:  
        # Меняем числа местами  
        from_num, to_num = to_num, from_num  
      
    # Используем формулу суммы арифметической прогрессии  
    n = to_num - from_num + 1  # количество элементов  
    total_sum = n * (from_num + to_num) // 2  
      
    return total_sum  
  
# Примеры использования:  
print(sum_distance(1, 5))   # Выведет 15 (1+2+3+4+5)  
print(sum_distance(5, 1))   # Выведет 15 (автоматически поменяет числа местами)  
print(sum_distance(3, 3))   # Выведет 3 (только одно число)  



```

```python
def trim_and_repeat(text: str, offset: int = 0, repetitions: int = 1) -> str:  
    """  
    Обрезает строку слева и повторяет её заданное количество раз.  
      
    :param text: исходная строка  
    :param offset: количество символов для обрезки слева (по умолчанию 0)  
    :param repetitions: количество повторений строки (по умолчанию 1)  
    :return: обработанная строка  
    """  
    # Проверяем корректность offset  
    if offset < 0:  
        raise ValueError("Параметр offset не может быть отрицательным")  
          
    # Обрезаем строку слева  
    trimmed_text = text[offset:]  
      
    # Повторяем строку нужное количество раз  
    result = trimmed_text * repetitions  
      
    return result  
  
# Примеры использования:  
# print(trim_and_repeat("hello"))  # -> hello  
# print(trim_and_repeat("hello", 2))  # -> llo  
# print(trim_and_repeat("hello", 1, 3))  # -> elloelloello  
```
